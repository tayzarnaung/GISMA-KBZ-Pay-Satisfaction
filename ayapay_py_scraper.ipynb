```python
from google_play_scraper import reviews, Sort

result, _ = reviews(
    'com.ayaplus.subscriber',  # Aya Pay's package name
    lang='en',            # 'en' or 'my' for Burmese reviews
    country='mm',         # Myanmar
    sort=Sort.NEWEST,     # or Sort.RATING, Sort.HELPFULNESS
    count=1000            # number of reviews to fetch
)
print(len(result))
```

```python
import pandas as pd
df = pd.DataFrame(result)
df.to_csv('aya_pay_reviews_mm.csv', index=False)
print("Saved:", len(df), "reviews")
```
